### Importando bibliotecas

```python
import pandas as pd
import numpy as np
import matplotlib as plt
import seaborn as sb
from sklearn.neighbors import KNeighborsClassifier
```

### Importando dados das flores de arquivo Excel

```python
flores = pd.read_excel("IrisDataset.xls")
flores.dropna()
```

### Plotando os relacionamentos de pares no conjunto de dados.

```python
sb.pairplot(flores,hue='variety')
```

### Definindo as features

```python
X = np.array(flores.drop('variety', 1))
Y = np.array(flores['variety'])
```

### Instanciando o classificador KNN para 5 vizinhos

```python
knn = KNeighborsClassifier(n_neighbors=5)
```

### Fazendo o treinamento

```python
knn.fit(X,Y)
```

### Executando o predict para classificar as flores

```python
print(knn.predict([[6.8,3.2,5.9,2.3]])[0])
```

```python
print(knn.predict([[4.4,0.3,1.3,0.2]])[0])
```

```python
print(knn.predict([[6, 2.7,3.9,1.2]])[0])
```
